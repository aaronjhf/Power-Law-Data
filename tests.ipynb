## Silverstein Solver

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
%%capture

evals  = np.logspace(1, -5, 200)
density = silverstein_density(evals , alpha = 1.5, q = 2, Lc = 1000**-1.5)
rho_dict = {}
for z, rho in zip(evals, density):
    rho_dict[z] = abs(rho)
zL, zR = end_pts(rho_dict)
```

```python
plt.loglog(evals, [abs(rho) for rho in density])
plt.vlines([zL, zR], ymin = 1e-5, ymax = 5e3, linestyle="--", color="tab:green")
```
